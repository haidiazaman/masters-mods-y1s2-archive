# Time-series prediction

In this notebook, we investigate a few methods for time-series prediction on the sunspot dataset.

Our goal is to see
- how we can estimate trends and use them to predict future values
- how we can also use auto-regressive models to achieve the same effect

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from typing import List

sns.set_theme()
np.random.seed(5206)
````

## Load and visualise dataset

**Sunspots dataset**

From wikipedia
>Sunspots are phenomena on the Sun's photosphere that appear as temporary spots that are darker than the surrounding areas.
>They are regions of reduced surface temperature caused by concentrations of magnetic flux that inhibit convection.
>Sunspots appear within active regions, usually in pairs of opposite magnetic polarity.
>Their number varies according to the approximately 11-year solar cycle.

![sunspot](https://upload.wikimedia.org/wikipedia/commons/thumb/f/f8/NASA%27s_SDO_Observes_Largest_Sunspot_of_the_Solar_Cycle_%2815430820129%29.jpg/800px-NASA%27s_SDO_Observes_Largest_Sunspot_of_the_Solar_Cycle_%2815430820129%29.jpg)

Let us load the sunspots dataset, which contains monthly mean measurements of total sunspot numbers from 1749 (wow!) to 2021.

````python
df = pd.read_csv("./data/sunspots.csv", index_col=0)
df.plot(x="Date", y="Monthly Mean Total Sunspot Number", rot=45)
````

We observe that there does appear to have some periodicities in the data. 

We are going to analyse a coarse-grained version of this dataset, where instead of monthly averages, we will look at yearly averages

````python
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Calculate the yearly average of the 'Monthly Mean Total Sunspot Number' column
df_yearly_average = df.groupby(df['Date'].dt.year)['Monthly Mean Total Sunspot Number'].mean()

# Convert the result to a DataFrame and reset the index
df_yearly_average = df_yearly_average.to_frame().reset_index()

# Rename the columns to 'Year' and 'Yearly Mean Total Sunspot Number'
df_yearly_average = df_yearly_average.rename(columns={'Date': 'Year', 'Monthly Mean Total Sunspot Number': 'Yearly Mean Total Sunspot Number'})

# Plot the yearly average
df_yearly_average.plot(x="Year", y="Yearly Mean Total Sunspot Number", rot=45)
````

For subsequent analysis, we split the data into training and test sets.

Unlike standard classification/regression datasets, we should not do a random split because we want to make sure that model fitted on historical data is able to predict future data.

Thus, we split the dataset according to
- Train: all data before and including 1950
- Test: all data after 1950

````python
df_train = df_yearly_average.loc[df_yearly_average['Year'] <= 1950]
df_test = df_yearly_average.loc[df_yearly_average['Year'] > 1950]
````

## Trend based model

We first analyse trend based models.

Since we suspect a periodic model, let us first try to detect its period.

Recall that as in the lectures, we assume the hypothesis
$$
    f(t,\theta) = \sum_{j=0}^{P-1}
    a_j \exp(j(2\pi i / P) t),
    \qquad
    \theta = (a, P)
$$

Period detection involves finding $P$, which can be obtained visually be plotting the *periodogram*, which computes
the absolute values of the Fourier coefficients of the data.

Fortunately, `scipy.signal` has a ready-made implementation.

````python
from scipy import signal
````

````python
# Extract the sunspot values from the DataFrame
sunspot_values = df_train['Yearly Mean Total Sunspot Number'].values

# Compute the periodogram
frequencies, power_spectrum = signal.periodogram(sunspot_values)

# Plot the periodogram
plt.plot(frequencies, power_spectrum)
plt.xlabel('Frequency')
plt.ylabel('Power Spectrum')
plt.title('Periodogram of Sunspot Data')
````

We can extract the period by the inversion of the frequency corresponding to the maximal power.

````python
max_power_spectrum_index = np.argmax(power_spectrum)
max_power_spectrum_frequency = frequencies[max_power_spectrum_index]
print(f'Max power frequency: {max_power_spectrum_frequency}')
print(f'Detected period: {1. / max_power_spectrum_frequency} years')
````

We set $P$ to the detected period.

````python
P = 1./max_power_spectrum_frequency
````

We prepare the training data. For simplicity we can shift the year count so that the beginning of the dataset is counted as year 0.

````python
first_year = df_train['Year'].values[0]
t_train = df_train['Year'].values - first_year
x_train = df_train['Yearly Mean Total Sunspot Number'].values
````

With the detected period, we can simply solve hte regression problem
$$
    \min_{a} 
    \sum_t
    \left\|
        x(t)
        -
        \sum_{j=0}^{P-1}
        a_j \exp(j(2\pi i / P) t) 
    \right\|^2
$$
which is just a linear regression problem.

````python
def generate_features(t: np.ndarray, period: float) -> np.ndarray:
    """
    Generate features for time series prediction.

    Parameters:
    t (np.ndarray): Time value.
    period (float): Period of the time series.

    Returns:
    np.ndarray: Array of generated features.
    """
    # Generate cosine features
    cos_features = np.array([np.cos(2 * np.pi * t * j / period) for j in range(int(period))]).T

    # Generate sine features
    sin_features = np.array([np.sin(2 * np.pi * t * j / period) for j in range(int(period))]).T

    # Concatenate cosine and sine features
    features = np.concatenate((cos_features, sin_features), axis=1)

    return features
````

````python
features_train = generate_features(t=t_train, period=P)
````

We now fit the model using `LinearRegression` from `sklearn.linear_model`.

````python
from sklearn.linear_model import LinearRegression
````

````python
reg = LinearRegression()
reg.fit(features_train, x_train)
````

Let us look at the prediction on the test dataset.

````python
# Inputs
t_test = df_test['Year'].values - df_yearly_average['Year'].values[0]
features_test = generate_features(t=t_test, period=P)

# Outputs
y_test = df_test['Yearly Mean Total Sunspot Number'].values

# Predicted outputs
y_test_pred = reg.predict(features_test)
````

Let us look at the results

````python
plt.plot(t_test + first_year, y_test, label="True")
plt.plot(t_test + first_year, y_test_pred, label="Predicted")
plt.xlabel("Year")
plt.ylabel("Yearly Average Number of Sunspots")
plt.legend()
````

We observe that this is a reasonable fit, the periodic trend is generally predicted correctly, but there is some systematic errors.

## Autoregressive models

Next, instead of trend-based models, let us try linear autoregressive models.

We will use the implementation in `statsmodels` (see [here](https://www.statsmodels.org/stable/index.html)).

Install it in your environment
```bash
pip install statsmodels
```

````python
from statsmodels.tsa.ar_model import AutoReg
````

Let us now fit a auto-regressive model of order $p$, i.e. ar(p) model to the data.

Important question: *how do we select p*?

We observe in the following that the results depends on $p$!

````python
# Fit autoregressive model
p = 5 # try p = 5, 50, 100
ar = AutoReg(df_train['Yearly Mean Total Sunspot Number'], lags=p)
ar_model = ar.fit()
````

Let us use this model to predict on the test dataset

````python
y_test_pred_ar = ar_model.predict(start=len(df_train), end=len(df_train) + len(df_test) - 1)
````

````python
plt.plot(t_test + first_year, y_test, label="True")
plt.plot(t_test + first_year, y_test_pred_ar, label="Predicted")
plt.xlabel("Year")
plt.ylabel("Yearly Average Number of Sunspots")
plt.legend()
````

Observe that the results depend on $p$!
- For $p$ too small, the fit will be bad
- For large $p$, instabilities may occur

### Choosing the order of the AR(p) model

We will use a simple validation approach by splitting the training data into two datasets
- training
- validation

We will perform ar(p) fitting on the training data, and then evaluate the mean-squared-error of the prediction on the validation data.

Note that this is the simpelst form of hyper-parameter tuning. For more robust choices, we should perform cross validation.
For time series data, this can be done via the routine `sklearn.model_selection.TimeSeriesSplit`.

````python
df_train_train, df_train_val = df_train.loc[df_train['Year'] <= 1900], df_train.loc[df_train['Year'] > 1900]
````

We now write a routine to perform the above steps
- training on the training set
- evaluting mse on the validation set
- return a set of `orders` and their respective `mse`

````python
def choose_best_order(train: np.ndarray, val: np.ndarray, orders: List[int]) -> List[float]:
    mse: List[float] = []

    # run through the orders, fit the model and calculate the MSE on the validation set
    for p in orders:
        model: AutoReg = AutoReg(train, lags=p)
        predicted: np.ndarray = model.fit().predict(start=len(train), end=len(train) + len(val) - 1)
        mse.append(np.mean((predicted - val) ** 2))

    return mse
````

````python
orders = np.arange(1, 61)  # choose the range of orders to consider
mse = choose_best_order(train=df_train_train['Yearly Mean Total Sunspot Number'], val=df_train_val['Yearly Mean Total Sunspot Number'], orders=orders)

plt.plot(orders, mse)
plt.xlabel('Orders')
plt.ylabel('Validation Mean Squared Error')

best_order = orders[np.argmin(mse)]
print(f'Best order: {best_order}')
plt.axvline(x=best_order, color='r', linestyle='--')
````

Let us now re-fit the model using the identified best order, and evaluate our predictions

````python
ar = AutoReg(df_train['Yearly Mean Total Sunspot Number'], lags=best_order)
ar_model = ar.fit()
````

````python
y_test_pred_ar = ar_model.predict(start=len(df_train), end=len(df_train) + len(df_test) - 1)
````

````python
plt.plot(t_test + first_year, y_test, label="True")
plt.plot(t_test + first_year, y_test_pred_ar, label="Predicted")
plt.xlabel("Year")
plt.ylabel("Yearly Average Number of Sunspots")
plt.legend()
````

There are also more general routines written for selection of autoregressive orders, for example, see `statsmodels.tsa.ar_model.ar_select_order` [here](https://www.statsmodels.org/dev/generated/statsmodels.tsa.ar_model.ar_select_order.html).
